```python
# Import required library
import pandas as pd
import os

# Verify file exists in the current directory
file_path = 'project.csv'
if not os.path.exists(file_path):
    print(f"Error: '{file_path}' not found in {os.getcwd()}")
    print("Please ensure 'project.csv' is in your Jupyter working directory.")
else:
    # Load the CSV file
    try:
        df = pd.read_csv(file_path, skiprows=2, skipfooter=30, engine='python')
        print("File imported successfully!")
        # Display first few rows
        print("\nFirst 5 rows of the dataset:")
        display(df.head())  # Use display for nicer output in Jupyter
    except Exception as e:
        print(f"Error loading file: {e}")

# Optional: List all files in the directory to confirm
print("\nFiles in current directory:")
print(os.listdir('.'))
```

```python
def filter_by_level(df, level='STATE'):
    return df[df['Region_Type'].str.contains(level, case=False)]
```

```python
import matplotlib.pyplot as plt
import os

os.makedirs('figures', exist_ok=True)

df_state = filter_by_level(df, 'STATE')
df_state_total = df_state[df_state['Total_Rural_Urban'] == 'Total']
top_10_states = df_state_total.sort_values('Population_Total', ascending=False).head(10)

plt.figure(figsize=(10, 5))
top_10_states.plot(x='Name', y='Population_Total', kind='bar', color='skyblue')
plt.title('Top 10 States by Population')
plt.xlabel('State')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('figures/bar_population_state.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import os

%matplotlib inline

os.makedirs('figures', exist_ok=True)

if 'df' not in globals():
    print("Error: 'df' not defined. Please run import and cleaning steps first.")
else:
    if 'filter_by_level' not in globals():
        def filter_by_level(df, level='STATE'):
            return df[df['Region_Type'].str.contains(level, case=False)]
    
    df_state = filter_by_level(df, 'STATE')
    
    df_state_total = df_state[df_state['Total_Rural_Urban'] == 'Total']
    top_5_states = df_state_total.sort_values('Population_Total', ascending=False).head(5)['Name'].tolist()
    
    df_top_states = df_state[df_state['Name'].isin(top_5_states) & df_state['Total_Rural_Urban'].isin(['Rural', 'Urban'])]
    
    pop_data = df_top_states.pivot_table(values='Population_Total', index='Name', columns='Total_Rural_Urban', aggfunc='sum', fill_value=0)
    
    # Debug: Check data
    print("Top 5 states:", top_5_states)
    print("Pivoted data:\n", pop_data)
    if pop_data.empty:
        print("Error: No rural/urban data found for top states. Check 'Total_Rural_Urban' values.")
    else:
        # Plot stacked bar chart
        plt.figure(figsize=(10, 6))
        pop_data.plot(kind='bar', stacked=True, color=['lightgreen', 'lightcoral'], ax=plt.gca())
        plt.title('Rural vs Urban Population (Top 5 States)')
        plt.xlabel('State')
        plt.ylabel('Population')
        plt.xticks(rotation=45)
        plt.legend(title='Area Type')
        plt.tight_layout()
        plt.savefig('figures/stacked_bar_rural_urban.png')
        plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import os

%matplotlib inline

os.makedirs('figures', exist_ok=True)

if 'df' not in globals() or 'filter_by_level' not in globals():
    print("Error: 'df' or 'filter_by_level' not defined. Please run import and cleaning steps first.")
else:
    df_district = filter_by_level(df, 'DISTRICT')
    df_district_total = df_district[df_district['Total_Rural_Urban'] == 'Total']
    
    print("Number of district rows:", len(df_district_total))
    if len(df_district_total) == 0:
        print("Error: No district data found. Check 'Region_Type' and 'Total_Rural_Urban' filters.")
    else:
        top_10_districts = df_district_total.sort_values('Population_Total', ascending=False).head(10)
        
        print("Top 10 districts:\n", top_10_districts[['Name', 'Population_Total']])
        
        plt.figure(figsize=(10, 5))
        sns.barplot(data=top_10_districts, x='Population_Total', y='Name', palette='Blues_d')
        plt.title('Top 10 Districts by Population')
        plt.xlabel('Population')
        plt.ylabel('District')
        plt.tight_layout()
        plt.savefig('figures/bar_top_districts.png')
        plt.show()
```

```python
import matplotlib.pyplot as plt
import os

os.makedirs('figures', exist_ok=True)

df_state = filter_by_level(df, 'STATE')
top_5_states = df_state[df_state['Total_Rural_Urban'] == 'Total'].sort_values('Population_per_sq_km', ascending=False).head(5)['Name']
df_top = df_state[df_state['Name'].isin(top_5_states) & df_state['Total_Rural_Urban'].isin(['Rural', 'Urban'])]

plt.figure(figsize=(12, 6))
for status in ['Rural', 'Urban']:
    subset = df_top[df_top['Total_Rural_Urban'] == status]
    plt.plot(subset['Name'], subset['Population_per_sq_km'], label=status, marker='o')
plt.title('Population Density: Rural vs Urban (Top 5 States)')
plt.xlabel('State')
plt.ylabel('Population per sq. km')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.savefig('figures/line_density.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import os

%matplotlib inline

os.makedirs('figures', exist_ok=True)

region_data = df[df['Total_Rural_Urban'] == 'Total']
pop_by_region = region_data.groupby('Region_Type')['Population_Total'].sum()

print("\nPopulation by region type:")
print(pop_by_region)

if not pop_by_region.empty:
    # Plot pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(pop_by_region, labels=pop_by_region.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
    plt.title('Population by Region Type')
    plt.axis('equal')
    plt.savefig('figures/pie_region_type.png')
    plt.show()
else:
    print("Error: No population data found for region types.")
```
